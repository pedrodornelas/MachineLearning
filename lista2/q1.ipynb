```python
import numpy as np
from scipy.stats import uniform, norm
import matplotlib.pyplot as plt

# p(C_{+1})
p1 = 0.6
# p(C_{-1})
p2 = 0.4

# p(x|C_{+1})
x1 = norm(2,1)
# p(x|C_{-1})
x2 = uniform(-2,4)  # [-2, -2+4]

# points
x = np.linspace(-4, 8, 1000)

# discriminant functions
g1 = x1.pdf(x)*p1
g2 = x2.pdf(x)*p2

plt.plot(x, g1, label='$g_{+1}(x)$')
plt.plot(x, g2, label='$g_{-1}(x)$')
plt.legend(loc='upper left')
plt.show()
```

```python
boundary = 0.6787

error = 0
aux = 0
for i in range(len(x)):
    # erro 1
    if x[i] <= boundary:
        error += g1[i] # g_{+1}
        aux += 1
    # erro 2
    elif x[i] <= 2:
        error += g2[i] # g_{-1}
        aux += 1
pError = error/aux

print('Error probability = '+str(pError))
```
